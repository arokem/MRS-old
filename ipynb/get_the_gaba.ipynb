```python
# Import the necessary modules: 
# These are 3rd party:
import array
import os
import scipy.stats as stats

# This is from nitime: 
import nitime as nt

# These are ours:
import MRS
import MRS.analysis as ana
import MRS.utils as ut
import MRS.files as fi
```

```python
# Point to some data that we have in here: 
data_path = os.path.join(MRS.__path__[0] , 'data')
data_files = ['mrs_P71680.7', 'mrs_P76288.7', 'pure_gaba_P64024.7', 'P58368.7', 'P11111.7']
# Read from the p-file:
data = fi.get_data(os.path.join(data_path, data_files[-3]))
```

```python
# Combine the data from the different coils, weighting each one by RMS, and aligning their phases:
w_data, w_supp_data = ana.coil_combine(data.squeeze())
w_data.shape, w_supp_data.shape
```

```python
# Perform spectral analysis on the data:
f_nonw, nonw_sig = ana.get_spectra(nt.TimeSeries(w_supp_data, sampling_rate=5000.0))

echo1 = nonw_sig[0]
echo2 = nonw_sig[0] - nonw_sig[1]
```

```python
# Convert from Hz to ppm, assuming that the water peak is at 4.7 ppm and 0 Hz and 1 ppm = 127.8 Hz: 
f_ppm = ut.freq_to_ppm(f_nonw)

# Find the array index for 0 ppm and for 4.5 ppm:
idx0 = np.argmin(np.abs(f_ppm - -0.7))
idx1 = np.argmin(np.abs(f_ppm - 4.3))
idx0, idx1
```

```python
# We'll use this to slice into our objects:
idx = slice(idx1, idx0)
```

```python
# Bootstrapping on the spectra (from each transient):
boots_on = []
boots_off = []
for i in range(100):
    b_idx = np.random.randint(0, echo1.shape[0], echo1.shape[0])
    boots_on.append(np.mean(echo2[b_idx],0))
    boots_off.append(np.mean(echo1[b_idx],0))    

boots_on = np.array(boots_on)
boots_off = np.array(boots_off)
```
